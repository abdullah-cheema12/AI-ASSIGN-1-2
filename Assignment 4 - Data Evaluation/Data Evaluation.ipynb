# Assignment 4: Data Analysis and Visualization Using Pandas

## Objective
The objective of this assignment is to perform data analysis and visualization using Pandas and explain the dataset in detail.


## Data Set Selection
For this assignment, we will use the "Iris" dataset from the seaborn library.


## Data Loading
Load the selected dataset into a Pandas DataFrame.


```python
# Data Loading
import pandas as pd
import seaborn as sns

# Load the Iris dataset from seaborn
data = sns.load_dataset('iris')
data.head()
```

## Data Exploration
Perform a detailed exploration of the dataset to understand its structure, features, and statistical summary.


```python
# Display basic information about the dataset
print("Basic information about the dataset:")
print(data.info())

# Display statistical summary of the dataset
print("\nStatistical summary of the dataset:")
print(data.describe())

# Display the structure of the dataset
print("\nShape of the dataset (rows, columns):", data.shape)

# Display the first few rows of the dataset
print("\nFirst few rows of the dataset:")
print(data.head())

# Display the column names
print("\nColumn names of the dataset:")
print(data.columns)
```

### Analysis:
- The dataset contains 150 entries with 5 columns: `sepal_length`, `sepal_width`, `petal_length`, `petal_width`, and `species`.
- There are no missing values in the dataset.
- The `species` column is categorical, while the other columns are numerical.
- Statistical summary shows that `sepal_length` ranges from 4.3 to 7.9, `sepal_width` from 2.0 to 4.4, `petal_length` from 1.0 to 6.9, and `petal_width` from 0.1 to 2.5.


## Data Cleaning
Clean the data by handling missing values, duplicates, and performing any necessary data transformations.


```python
# Check for missing values
data.isnull().sum()

# Check for duplicates
data.duplicated().sum()

# If there are any missing values or duplicates
data = data.dropna()  # Drop missing values
data = data.drop_duplicates()  # Drop duplicate rows
```

### Analysis:
- The dataset does not contain any missing values.
- There are no duplicate rows in the dataset.
- The dataset is clean and ready for further analysis and visualization.


## Data Visualization
Use Pandas built-in visualization capabilities along with Matplotlib and Seaborn to create various graphs and charts.
### Pairplot
The pairplot shows pairwise relationships in the dataset, allowing us to see how the features are related to each other for different species.


```python
# Pairplot to visualize pairwise relationships in the dataset
import matplotlib.pyplot as plt

sns.pairplot(data, hue='species')
plt.show()
```

### Analysis:
- The pairplot shows that petal length and petal width are highly correlated, particularly for the species `setosa`.
- Sepal length and sepal width are less correlated.
- This visualization helps in understanding the distribution and relationship between different features for each species.


### Boxplot
The boxplot shows the distribution of each feature in the dataset.


```python
# Boxplot to visualize the distribution of each feature
plt.figure(figsize=(10, 6))
sns.boxplot(data=data)
plt.title('Boxplot of Iris Dataset Features')
plt.show()
```

### Analysis:
- The boxplot indicates that the `petal_length` and `petal_width` have more variation compared to `sepal_length` and `sepal_width`.
- There are some outliers in the `sepal_width`.
- This visualization helps in understanding the range and distribution of the dataset features.


### Histogram
The histograms show the distribution of each feature.


```python
# Histogram to visualize the distribution of each feature
data.hist(bins=20, figsize=(10, 6))
plt.suptitle('Histogram of Iris Dataset Features')
plt.show()
```

### Analysis:
- The histograms show that the `petal_length` and `petal_width` have a bimodal distribution, indicating two distinct groups.
- `Sepal_length` and `sepal_width` are more normally distributed.
- This visualization helps in understanding the frequency distribution of the dataset features.

### Heatmap
The heatmap shows the correlation matrix of the features.


```python
# Heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 6))
numeric_data = data.drop(columns=['species'])  # Exclude categorical column
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Iris Dataset (Excluding Species)')
plt.show()
```

### Analysis:
- The heatmap shows a high positive correlation between `petal_length` and `petal_width`.
- `Sepal_width` has a weak correlation with other features.
- This visualization helps in understanding the strength and direction of relationships between different features.
